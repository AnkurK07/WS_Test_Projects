```python
# Importing required Library
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#### <span style="color:red">In this project my aim is to extrct all the name of movies,Genres and IMDB ratings  wchich is already released or Yet to be release so let start</span>

```python
# Scrapping movies which is already released or yet to be release from IMDB
url1='https://www.imdb.com/search/title/?title_type=feature&year=2023-01-01,2023-12-31'
r1=requests.get(url1)
soup2=BeautifulSoup(r1.text,'lxml')

# I an not Showing here Scrapped page because it is very lengthy
```

```python
# Now Extracting Movies Name 
movies=soup2.select("h3.lister-item-header")
Movies_name=[i.text.strip() for i in movies]
MN=[j.split('\n') for j in Movies_name]
Name_of_Movies=[i[1] for i in MN]
len(Name_of_Movies)
Name_of_Movies
```

```python
# Now getting the Genres
Gnr=soup2.select('span.genre')
Gnr_1=[i.text for i in Gnr]
Gnr_2=[i.split('\n') for i in Gnr_1]
Gnr_3=[j[1] for j in Gnr_2]
Gnr_4=[k.split('     ') for k in Gnr_3]
Gnr_5=[l[0] for l in Gnr_4]
len(Gnr_5)
Gnr_5
```

```python
# Now Extracting the Ratings
Ratings_1=soup2.select('strong')
X=['Detailed','Popularity']
Ratings_2=[i.text for i in Ratings_1]
Ratings_3=[float(i) for i in Ratings_2 if i not in X]
len(Ratings_3)
Ratings_3
```

#### <span style="color:blue">Now there is a problem that is the movies yet to be release is not rated yet and i did't get the ratings so my array number is not the same and i can't make data frame with it.</span>

```python
# Shorting the Movies which are not rated yet. (if my dada is too big then i can't do this step!)
NR=['Barbie','Oppenheimer','Killers of the Flower Moon',"Dune: Part Two","Five Nights at Freddy's",'Red, White & Royal Blue','Rebel Moon','Haunted Mansion','Teenage Mutant Ninja Turtles: Mutant Mayhem','The Nun II']
NRdf=pd.DataFrame({'Name of Movies in 2023':NR,'Ratings':['NR' for i in range(10) ]})
print(len(NRdf))
NRdf
```

```python
# Sorting the Rated Movies
RT=[i for i in Name_of_Movies if i not in NR ]
RTdf=pd.DataFrame({'Name of Movies in 2023':RT,'Ratings':Ratings_3})
RTdf
```

```python
# Now making a dataframe without rating. i will merge with Ratings later
IMdb=pd.DataFrame({'Name of Movies in 2023':Name_of_Movies ,'Genres':Gnr_5})
IMdb
```

```python
# Now Joining Two Tables which are rated and unrated
Ndf=pd.concat([RTdf,NRdf],axis=0) # Merging Both Data Frame
Ndf=Ndf.reset_index(drop=True)     # Reset My Index
Ndf
```

```python
# Now add it my Data Frame By join() Statement
Final_df=pd.merge(IMdb,Ndf,how='outer')
Final_df
```

```python
# Now Making It CSV File
Final_df.to_csv('Final_df.csv')
```
